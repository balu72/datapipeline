```python
import pandas as pd

# Ingestion Layer
def ingest_data():
    file_path = '/Users/balupillai/BITS Digital/LiveSession/handson/LiverPatientDataset_train.csv'
    data = pd.read_csv(file_path, encoding='latin1')
    #print (data)
    return data
```

```python
# Preprocessing Layer
def preprocess_data(data):
    # Clean data
    # Print count of records
    print("Total records:", len(data))
    print("Number of records with Total Billirubin missing-Before ", len(data[data['Total Bilirubin'].isna()]))
    data = data.dropna(subset=['Total Bilirubin'])
    print("Number of records with Total Billirubin missing- After", len(data[data['Total Bilirubin'].isna()]))

    # Print count of records
    print("Total records:", len(data))

    return data
```

```python
def validate_data(data):
    # Validate Bilirubin's value, more than 10 is invalid
    
    print("Total records:", len(data))
    count_ge_10 = len(data[data['Total Bilirubin'] >= 10])
    count_lt_10 = len(data[data['Total Bilirubin'] < 10])

    print("Number of records with Total Bilirubin >= 10:", count_ge_10)
    print("Number of records with Total Bilirubin < 10:", count_lt_10)

    # Drop invalid records
    data = data[data['Total Bilirubin'] < 10]
    
    # Print count of records
    print("Total records:", len(data))
```

```python
# Main Execution
def main():
    data = ingest_data()
    preprocessed_data = preprocess_data(data)
    validate_data(preprocessed_data)

# Example usage
main()
```

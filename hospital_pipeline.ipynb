```python
import pandas as pd

# Ingestion Layer
def ingest_data(file_path):
    data = pd.read_csv(file_path)
    return data
```

```python
# Preprocessing Layer
def preprocess_data(data):
    # Clean data
    data = data.dropna()  # Remove missing values
    data = data.drop_duplicates()  # Remove duplicates
    
    # Transform data (example: normalize a column)
    if 'value' in data.columns:
        data['value'] = (data['value'] - data['value'].mean()) / data['value'].std()
    
    return data
```

```python
# Validation Layer
def validate_data(data):
    if not all(data.dtypes == 'float64'):
        raise ValueError("Data validation failed: Not all columns are float64")
    if data['value'].min() < 0 or data['value'].max() > 1:
        raise ValueError("Data validation failed: 'value' out of range [0, 1]")
    print("Data validation passed.")
```

```python
# Main Execution
def main(file_path):
    data = ingest_data(file_path)
    preprocessed_data = preprocess_data(data)
    validate_data(preprocessed_data)

# Example usage
# main('path/to/your/data.csv')
```
